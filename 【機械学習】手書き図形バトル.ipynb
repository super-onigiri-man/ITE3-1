```python
# ファイル名を指定した画像を NumPy配列 にして返す関数
def img2array(img_path):
    img_c1 = Image.open(img_path).convert("L")
#    plt.imshow(img_c1, cmap="gray")
#    plt.show()
    img_c2 = img_c1.resize((8,8),Image.LANCZOS)
#    plt.imshow(img_c2, cmap="gray")
#    plt.show()
    img_arr1 = np.asarray(img_c2, dtype=float)
#    print(img_arr1)
    img_arr2 = 16 - np.floor(17 * img_arr1/256)
#    print(img_arr2)
#    plt.imshow(img_arr2, cmap="gray")
#    plt.show()
    img_arr3 = img_arr2.reshape(1,-1)
#    print(img_arr3.shape)
    return img_arr3
```

```python
def imgs2csv(img_list, csv_name):
    csv_array = np.empty((0, 64+1),float)
    for img_path in img_list:
        #print(img2array(img_path))
        ans_label = -1
        if re.search('_maru', img_path):
            ans_label = 0
        elif re.search('_batsu', img_path):
            ans_label = 1
        elif re.search('_batu',img_path):
            ans_label = 1
        elif re.search('_sankaku', img_path):
            ans_label = 2
        elif re.search('_shikaku', img_path):
            ans_label = 3
        elif re.search('_sikaku',img_path):
            ans_label = 3

        img_arr_x = img2array(img_path)
        img_arr_xy = np.append(img_arr_x, np.array([[ans_label]]), axis=1)

        csv_array = np.append(csv_array, img_arr_xy,  axis=0)

    h_str = ",".join(["x"+str(i) for i in range(1,65)]) + ",y(label)"
    np.savetxt(csv_name, csv_array, delimiter=",", comments="", header=h_str)
    return csv_array
```

```python
def folder2csv(dir_path, csv_name, ex_list):
    file_list = glob.glob(dir_path + '*.*')
    # dir_path + '*.*' と打つと、拡張子があるファイルのみがリスト化される。
    img_list = []

    for file_name in file_list:
        name, ext = os.path.splitext(file_name)
        #print(ext)
        if ext in ex_list:
            img_list.append(file_name)
    print("あなたが指定したフォルダ内の画像リストは")
    print(img_list)
    img_array = imgs2csv(img_list, csv_name)
    return img_array
```

```python
from skopt import BayesSearchCV
from sklearn.svm import SVR
from sklearn.datasets import load_digits

# ハイパーパラメータ探索範囲の定義
param_space = {
    'C': (1e-6, 1e+6, 'log-uniform'),
    'gamma': (1e-6, 1e+1, 'log-uniform'),
    'degree': (1, 8),  # integer valued parameter
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
}

# ベイズ最適化の実行
opt = BayesSearchCV(SVR(), param_space, n_iter=50, cv=5, n_jobs=-1)
opt.fit(X, y)

# 最適なハイパーパラメータの表示
print("Best parameters found: ", opt.best_params_)
```

```python
# ハイパーパラメータ探索範囲の定義
param_space = {
    'n_neighbors': (1, 10),       # 近傍点の数
    'weights': ['uniform', 'distance'],  # 近傍点の重みづけ
    'p': (1, 2),                  # ミンコフスキー距離のパラメータ (1: マンハッタン距離, 2: ユークリッド距離)
}

# ベイズ最適化の実行
opt = BayesSearchCV(
    KNeighborsClassifier(),
    param_space,
    n_iter=50,
    cv=5,
    n_jobs=-1
)
opt.fit(X_train, y_train)

# 最適なハイパーパラメータの表示
print("Best parameters found: ", opt.best_params_)

# 最適なハイパーパラメータを使用してモデルを構築
best_params = opt.best_params_
knn_model = KNeighborsClassifier(
    n_neighbors=best_params['n_neighbors'],
    weights=best_params['weights'],
    p=best_params['p']
)

# モデルの評価
accuracy = knn_model.fit(X_train, y_train).score(X_test, y_test)
print("Test accuracy with best parameters:", accuracy)
```

```python
# アルゴリズム1: ランダムフォレスト
rf_model = RandomForestClassifier(n_estimators=1000, random_state=3)
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)

# アルゴリズム2: サポートベクトルマシン
svm_model = SVC( C=1.0, random_state=3)
svm_model.fit(X_train, y_train)
svm_predictions = svm_model.predict(X_test)

# 予測結果を組み合わせる（例：単純な多数決）
ensemble_predictions = (rf_predictions + svm_predictions) // 2

# 組み合わせたモデルの精度を評価
ensemble_accuracy = accuracy_score(y_test, ensemble_predictions)
print(f'Ensemble Model Accuracy: {ensemble_accuracy}')
```

```python


from keras.preprocessing.image import load_img, img_to_array
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
import os
import glob

# 1枚あたりN_img枚の画像を水増し
N_img = 5

# 入力画像の保存先パス
input_path = "./pngshare/"
files = glob.glob(input_path + '/*.png')

# 出力画像の保存先パス
output_path = "./png_sharesum1128/"
if os.path.isdir(output_path) == False:
    os.mkdir(output_path)


for i, file in enumerate(files):

    img = load_img(file)
    x = img_to_array(img)
    x = np.expand_dims(x, axis=0)


    params = {
    # 'zoom_range': [0.5, 1.0], #拡大・縮小
    'rotation_range': 45, #回転
    'vertical_flip': True, #上下反転（ランダム）
    'horizontal_flip': True, #左右反転（ランダム）
    'height_shift_range': 0.1, #位置変更（y軸方向）
    'width_shift_range': 0.1, #位置変更(垂直)
    # 'shear_range': 90,　#せん断
    # 'channel_shift_range': 127, #色素
    # 'fill_mode': 'constant', #これを実行すると周りが黒くなる
    }


    # ImageDataGeneratorの生成
    datagen = ImageDataGenerator(**params)



    # ファイル名から拡張子を取り除く
    filename = os.path.splitext(os.path.basename(file))[0]

    # 1枚あたり20枚の画像を水増し生成
    dg = datagen.flow(x, batch_size=1, save_to_dir=output_path, save_prefix=filename, save_format='png')
    for i in range(N_img):
        batch = dg.next()
```
